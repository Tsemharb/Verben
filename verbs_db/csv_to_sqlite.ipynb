```python
import csv
import sqlite3
sqlite_file = 'verbs_db.sqlite'
table_name = 'table'
id_column = 'ID'
de_column = 'de'
ru_column = 'ru'
status_column = 'status'
timing_column = 'timing'
```

```python
with open ('500.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=';')
    for row in reader:
        print(row)
```

```python
conn = sqlite3.connect(sqlite_file)
c = conn.cursor()
```

```python
c.execute('CREATE TABLE tab (ID INTEGER NOT NULL PRIMARY KEY,\
                            de TEXT,\
                            ru TEXT,\
                            status INTEGER,\
                            timing INTEGER)')
```

```python
with open ('500.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=';')
    for row in reader:
        c.execute("INSERT INTO tab (de, ru, status, timing) VALUES ('{de}','{ru}', 0, 0)"\
                 .format(de=row[0], ru=row[1]))

conn.commit()
conn.close()
```
